# 2019/01/15 test campaign 1, day 2
##    3.4 Measure M1 various burst sizes; measure M1-M3-M1 with best burst size

```python
from M1M3tools import *
import numpy as np
import matplotlib.pyplot as plt
import glob

%matplotlib inline
plt.jet()

dateID = 190115
testID = '3-4'
iter = ' Bad M3'
```

## First we wanted to optimize the number of samples and time interval for each measurement
We did this test with M1 only

### Measurement 1 : 5 samples separated by 10 seconds

```python
fig, ax = plt.subplots(2,2,figsize=(15,10))
i = 0
for iset in range(1,5):
    irow = np.int8(i/2)
    icol = np.mod(i,2)
    i += 1
    print('%s set %s ------ irow = %d, icol = %d, -----------------------------'%(testID, iset, irow, icol))
    pattern = '%s/Processed/%d_%s%s/%d %s M1 burst5x10s set%d morph.h5'%(
        dataDir, dateID, testID, iter, dateID, testID, iset)
    fileset = sorted(glob.glob(pattern))
    if len(fileset) == 0:
        print(pattern)
    m1s,centerRow1,centerCol1,pixelSize1 = readH5Map(fileset)
    if iset==1:
        m1s1 = m1s.copy()
    elif iset == 2:
        m1s2 = m1s.copy()
    elif iset==3:
        m1s3 = m1s.copy()
    elif iset==4:
        m1s4 = m1s.copy()
    m1rms = np.std(m1s[~np.isnan(m1s)])*1000 #convert into nm
    img = ax[irow][icol].imshow(m1s, origin='lowerleft')
    fig.colorbar(img, ax=ax[irow][icol])
    ax[irow][icol].set_title('%s%s set%d, RMS_M1=%.0f nm'%(testID,iter, iset, m1rms))
avg = (m1s1+m1s2+m1s3+m1s4)/4
idx = ~np.isnan(avg)
rms1 = np.std(np.hstack((m1s1[idx] -avg[idx], m1s2[idx]-avg[idx], m1s3[idx]-avg[idx], m1s3[idx]-avg[idx])))    
```
